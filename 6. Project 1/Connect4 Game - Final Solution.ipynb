# Connect4 Game - Final Solution

Here is the final solution to Project 1 of the course (Connect4 Game).

```python
# GLOBAL VARIABLES

# integers represent the number of rows and columns
NUMROW = 5
NUMCOL = 5

# strings represent the x winner and o winner
XWINNER = "XXXX"
OWINNER = "OOOO"


# create connect4 board
def create_board(row, column) :
    board = [["*" for i in range(column)] for j in range(row)]
    return board

# display connect4 board on the screen
def print_board(board):
    for num in range(0, NUMCOL):
        print(" " + str(num), end='')
    print()

    for row in range(0, NUMROW):
        for col in range(0, NUMCOL):
            print(" " + str(board[row][col]), end='')
        print()


# check horizontal match
def find_horizontal_match(board):
    win_flag = False
    
    row_str_arr = []
    for row in range(0, NUMROW):
        str = ""
        for col in range(0, NUMCOL):
            str = str + board[row][col]
        row_str_arr.append(str)

    for row_str in row_str_arr:
        if XWINNER in row_str or OWINNER in row_str:
            win_flag = True
            break
    return win_flag

# check vertical match
def find_vertical_match(board):
    win_flag = False
    
    col_str_arr = []
    for col in range(0, NUMCOL):
        str = ""
        for row in range(0, NUMROW):
            str = str + board[row][col]
        col_str_arr.append(str)

    for col_str in col_str_arr:
        if XWINNER in col_str or OWINNER in col_str :
            win_flag = True
            break

    return win_flag

# check diagonal match
def find_diagonal_match(board):
    win_flag = False
    
    diag_str_list = [''] * (NUMROW*NUMCOL)
    for col in range(0, NUMCOL):
        diag_num = col

        for row in range(0, NUMROW):
            if row + col > (NUMROW-1) :  # dont go beyond right bound of board
                break
            if col == 0:  
                diag_str_list[diag_num] = diag_str_list[diag_num] + board[row][row]  # middle
            else :
                diag_str_list[diag_num] = diag_str_list[diag_num] + board[row][row + col]  # upper
                diag_str_list[diag_num + (NUMROW-1)] = diag_str_list[diag_num + (NUMROW-1)] + board[row + col][row]  # lower

    # get diagonal match
    for col in range(0, NUMCOL):
        diag_num = col + (NUMROW+NUMCOL)

        for row in range(0, NUMROW):
            row_id = (NUMROW-1) - row  # to start from right to left
            if row_id - col < 0 :  # dont go beyond left bound of board
                break

            if col == 0:
                diag_str_list[diag_num] = diag_str_list[diag_num] + board[row][row_id]  # middle
            else :
                diag_str_list[diag_num] = diag_str_list[diag_num] + board[row][row_id - col]  # upper
                diag_str_list[diag_num + (NUMROW-1)] = diag_str_list[diag_num + (NUMROW-1)] + board[row + col][row_id]  # lower

    for diag_str in diag_str_list:
        if XWINNER in diag_str or OWINNER in diag_str:
            win_flag = True
            break

    return win_flag


def play_connect4():
    
    print("Welcome to Connect4 Game!")
    
    board = create_board(NUMROW, NUMCOL)
    print_board(board)

    turn = 0
    while True:
        player = 'A'
        if turn % 2 == 0 :
            player = 'A'
        else :
            player = 'B'

        print(".................................................. ")
        print("Player", player , " turn >>>>>......................... ")
        checker = str(input("Please select X or O >> "))
        if checker != 'X' and checker != 'O' :
            print("Invalid checker", checker)
            continue

        print("Please enter column number 0 -", NUMCOL, ">> ")
        col_num = int(input())
        if col_num > NUMCOL or col_num < 0:
            print("Wrong entry -> Please enter column number 0 -", NUMCOL, ">> ")
            continue

        for row in range(NUMROW-1, -1, -1):
            if board[row][col_num] == "*" :
                board[row][col_num] = checker
                break;

        turn += 1
        print_board(board)
        
        if find_horizontal_match(board) or find_vertical_match(board) or find_diagonal_match(board):
            print(".................................................. ")
            print("Player : ", player , " Win the match by ", turn, " Steps")
            print(".................................................. ")
            break

        if turn >= 20:
            print("You have crossed maximum limits. Please start again.")
            break

play_connect4()
```
