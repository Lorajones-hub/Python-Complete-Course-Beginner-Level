In this project you will create a simple game in Python called connect 4. It is a variation of Tic-Tac_Toe game. The rules are simple:

- Try to build a row of four checkers while keeping your opponent from doing the same.
- There are 2 players in this game and the checkers are either the character'X' or 'O'.
- The winner is the first player to get 4 colored checkers in a row $-$ horizontally, vertically, or diagonally.
- This is a grid (or borad) game, so you need to use a 2 dimentional matrix of any size to implement the grid.
- Players have to alternate turns. Each palyer has to choose a column to add his checker vertically on the board. The vertical strategy creates a unique challenge.
